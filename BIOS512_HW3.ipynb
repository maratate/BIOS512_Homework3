### Question 1
Imagine we have a data frame called `data`, with a `type` column. Which one works and why?\
Function 1:
```
group_and_tally <- function(df, column){
    df %>% group_by({{ column }}) %>% tally();
}
group_and_tally(data, type);
```

Function 2:
```
group_and_tally <- function(df, column){
    df %>% group_by(column) %>% tally();
}
group_and_tally(data, type);
```
## Answer: 
Function 1 works because of tidy evaluation to correctly pass column names. In function 1, the syntax uses curly brackets {{}} tells group_by() to read it as a type, not a literal object called “column”. Function 2 does not evaluate column as a variable and tries to group by a column named "column", instead of the value passed into the function.

### Question 5

Download the patient_names.csv and patient_properties.csv files from Canvas and read them into R. Manually set the date columns to be date variables. Print the first 10 observations of each.

````r
library(dplyr)
library(lubridate)

df_names <- read_csv("~/Desktop/BIOS512/Homework3/patient_names.csv")
df_properties <- read_csv("~/Desktop/BIOS512/Homework3/patient_properties.csv")

df_names <- df_names %>%
  mutate(
    BIRTHDATE = mdy(BIRTHDATE),
    DEATHDATE = mdy(DEATHDATE),
    BIRTHDATE = if_else(
      BIRTHDATE > as.Date("2020-01-01"), 
      BIRTHDATE - years(100),
      BIRTHDATE
    )
  )

print(df_names, n = 10)
print(df_properties, n = 10)
````

### Question 6
In the data frame pulled from patient_properties, you'll notice that the data is long, not wide. Do a pivot to make the properties their own columns. Print the first 10 observations after you do so.

````r
df_properties_wide <- df_properties %>%
  pivot_wider(
    names_from = property,
    values_from = value
    )
print(df_properties_wide, n = 10)
````

### Question 7
Perform a left join of the names and properties_wide data frames by the ID column 
and print the first 10 rows.

````r
df_joined <- df_names %>%
  left_join(df_properties_wide, by = "ID")

print(df_joined, n = 10)
````

### Question 8
Notice something interesting about the names in our data set. Fix the name formatting and print the first 10 observations.

````r
df_joined <- df_joined %>%
  mutate(
    FIRST = str_replace_all(FIRST, "[^A-Za-z\\s'-]", ""),
    LAST = str_replace_all(LAST, "[^A-Za-z\\s'-]", "")
  )

print(df_joined, n = 10)
````

### Question 9
Using a for statement to loop through the categorical variables (excluding name and ID), print the counts of each unique value in descending order, using the mdpre() function for formatting.

````r
for (col in c("MARITAL", "RACE", "ETHNICITY", "GENDER", "STATE", "CITY")) {
  cat("###", col, "\n") 
  df_counts <- df_joined %>%
    count(.data[[col]]) %>%
    arrange(desc(n))
  mdpre(df_counts)
}
````

### Question 10
If you see any weird values, get rid of the ones that don't make sense, and combine the ones that are formatted wrong. Don't forget to check the dates! Print the new tables for categorical values, and print the date ranges.

````r

# Cleaning up MARITAL
df_filter <- df_joined %>%
  mutate(
    MARITAL = case_when(
      str_to_upper(MARITAL) %in% c("M") ~ "Married",
      str_to_upper(MARITAL) %in% c("S") ~ "Single",
      TRUE ~ NA_character_  
    )
  )

#Cleaning up RACE

df_filter<- df_filter %>%
  mutate(
    RACE = case_when(
      RACE == "asiann" ~ "asian",
      TRUE ~ RACE  
    )
  )

#Cleaning up Ethnicity

df_filter <- df_filter %>%
  mutate(
    ETHNICITY = case_when(
      str_to_upper(ETHNICITY) %in% c("HISPANI", "HISPANIC") ~ "Hispanic",
      str_to_upper(ETHNICITY) %in% c("NONHISPANI", "NONHISPANIC") ~ "Non-Hispanic",
      TRUE ~ NA_character_
    )
  )

#Cleaning up Gender

df_filter <- df_filter %>%
  mutate(GENDER = case_when(
    tolower(GENDER) == "female" ~ "F",
    tolower(GENDER) == "male" ~ "M",
    TRUE ~ GENDER  
  ))

#Printing new tables for categorical variables 
for (col in c("MARITAL", "RACE", "ETHNICITY", "GENDER", "STATE", "CITY")) {
  cat("###", col, "\n") 
  df_counts2 <- df_filter %>%
    count(.data[[col]]) %>%
    arrange(desc(n))
  mdpre(df_counts2)
}

#Printing date ranges 
birthdate_range <- range(df_filter$BIRTHDATE, na.rm = TRUE)
deathdate_range <- range(df_filter$DEATHDATE, na.rm = TRUE)

cat("BIRTHDATE Range:", format(birthdate_range[1]), "to", format(birthdate_range[2]), "\n")
cat("DEATHDATE Range:", format(deathdate_range[1]), "to", format(deathdate_range[2]), "\n")
````

### Question 11
Make a histogram of the ages of patients by gender. 

````r
#Calculate and create an Age variable (Assuming NA DEATHDATES mean they are still alive)

df_age <- df_filter %>%
  mutate(AGE = if_else(
    !is.na(DEATHDATE) & DEATHDATE >= BIRTHDATE,
    as.numeric(DEATHDATE - BIRTHDATE) / 365.25,
    as.numeric(Sys.Date() - BIRTHDATE) / 365.25
  ))

#Make the histogram of age by gender 

library(ggplot2)

ggplot(df_age, aes(x=AGE, fill=GENDER)) +
  geom_histogram(position = "dodge", binwidth = 5) +
  labs(title = "Distribution of Age by Gender",
       x = "Age (years)",
       y = "Number of Patients") +
  theme_minimal()
````

### Question 12
Make a scatterplot of birthdate by martial status.

````r
ggplot(df_age, aes(x = AGE, y = MARITAL, color = MARITAL)) +
  geom_jitter(width = 0.2, alpha = 0.6) +
  labs(
    x = "Age (years)",
    y = "Marital Status",
    title = "Patient's Age by Marital Status"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
````
